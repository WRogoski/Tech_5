###Ajustando DF

```python
df_2 = df[['NOME', 'ANOS_PM_2020', 'FASE_TURMA_2020', 'FASE_2021', 'FASE_2022', 'IAA_2020', 'IAA_2021', 'IAA_2022', 'IAN_2020', 'IAN_2021', 'IAN_2022', 'IDA_2020', 'IDA_2021', 'IDA_2022', 'IEG_2020', 'IEG_2021', 'IEG_2022', 'INDE_2020', 'INDE_2021', 'INDE_2022', 'IPP_2020', 'IPP_2021', 'IPP_2022', 'IPS_2020', 'IPS_2021', 'IPS_2022', 'IPV_2020', 'IPV_2021', 'IPV_2022', 'PEDRA_2020', 'PEDRA_2021', 'PEDRA_2022', 'PONTO_VIRADA_2020', 'PONTO_VIRADA_2021', 'PONTO_VIRADA_2022']]


df_2.head()
```

```python
base_alunos = 'https://raw.githubusercontent.com/WRogoski/Tech_5/main/TbAluno.csv'
bd_alunos = pd.read_csv(base_alunos)
bd_alunos.head()
```

```python
bd_alunos['NomeAluno'] = bd_alunos['NomeAluno'].str.replace('Aluno ', 'ALUNO-', regex=False)
bd_alunos.head()
```

```python
bd_alunos_2 = bd_alunos[['NomeAluno', 'Sexo', 'DataNascimento', 'CorRaca']]
bd_alunos_2.head()
```

```python
bd_alunos_2.info()
```

```python
bd_alunos_2.rename(columns={'NomeAluno': 'NOME'}, inplace=True)
```

```python
bd_alunos_2.head()
```

```python
df_3 = pd.merge(df_2, bd_alunos_2, on='NOME')
df_3.head()
```

```python
df_3.shape
```

```python
df_3.info()
```

```python
df_3['FASE_TURMA_2020'] = df_3['FASE_TURMA_2020'].str.extract('(\w)', expand=False)

df_3.head()
```

```python
converter = ['ANOS_PM_2020','FASE_TURMA_2020','IAA_2020','IAA_2022','IAN_2020','IDA_2020','IDA_2022','IEG_2020','IEG_2022','INDE_2020','INDE_2021','INDE_2022','IPP_2020','IPP_2022','IPS_2020','IPV_2020','IPV_2022']
```

```python
for col in converter:
    try:
        df_3[col] = pd.to_numeric(df_3[col], errors='coerce')
    except:
        pass

# Verificando os tipos de dados após a conversão
print(df_3.dtypes)
```

```python
df_3['DataNascimento'] = pd.to_datetime(df_3['DataNascimento'], errors='coerce')
df_3['DataNascimento'] = df_3['DataNascimento'].dt.strftime('%d/%m/%Y')
df_3.head()
```
